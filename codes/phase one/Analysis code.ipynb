```python
import pandas as pd
```

```python
#Load data
page_events = pd.read_csv("View and shopping.csv")
```

```python
#Get all unique session ids in the search_events dataset
sess_id = list(page_events.session_id.unique())
sess_id
```

```python
#new dataframe to hold results
sess_page = pd.DataFrame(columns=["session_id", "page_title"])
page = pd.DataFrame(columns=["event_type", "event_date"])
```

```python
for sess_id in sess_id[3:9]:
    pages_list = list(page_events[page_events["session_id"] == sess_id]["page_title"]) 
    type_list = list(page_events[page_events["session_id"] == sess_id]["event_type"]) 
    date_list = list(page_events[page_events["session_id"] == sess_id]["event_date"])
    pages = ",".join(pages_list)
    types = type_list
    dates = ",".join(date_list)
    sess_page.loc[-1] = [sess_id, pages]
    sess_page.index += 1
    page.loc[-1] = [types, dates]
    page.index += 1

table = pd.concat([sess_page, page], axis=1)
table
```

```python
table.to_csv('Trial.csv', index=False)
```
