```python
# _
import pandas as pd
```

```python
orders_file_path = "../../../Data/DEngage Data/order_event/cb3e4c81-8b19-4025-836e-2fad364f8b38.csv"
page_file_path = "../../../Data/DEngage Data/page_view_event/913e9b97-a510-41ec-96e4-7a3ba13de494.csv"

page_view = pd.read_csv(page_file_path)
orders = pd.read_csv(orders_file_path)
```

```python
order_sess_ids = list(orders["session_id"])

sess_ids = page_view["session_id"]
sess_ids = list(set(sess_ids)) #unique session ids

click_pattern = pd.DataFrame(columns=["session_id", "page_urls", "page_types", "page_titles", "order"])

for sess_id in sess_ids:
    print(sess_id)
    temp = page_view[page_view["session_id"] == sess_id]
    page_urls = "->".join(list(temp["page_url"].astype(str)))
    page_types = "->".join(list(temp["page_type"].astype(str)))
    page_titles = "->".join(list(temp["page_title"].astype(str)))
    order = "No"
    if sess_id in order_sess_ids:
        order = "Yes"
        
    click_pattern.loc[-1] = [sess_id, page_urls, page_types, page_titles, order]
    click_pattern.index += 1
```

```python
click_pattern.to_csv("click_pattern")
click_pattern[click_pattern["order"] == "No"]
```
